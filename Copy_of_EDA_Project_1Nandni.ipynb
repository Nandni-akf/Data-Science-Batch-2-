# **Cleaning of the whole data**

````python
import pandas as pd

# Load dataset
cars_df = pd.read_csv("quikr_car.csv")

# -------------------------------
# Clean Year
# -------------------------------
cars_df = cars_df[cars_df['year'].astype(str).str.isnumeric()]
cars_df['year'] = cars_df['year'].astype(int)

# -------------------------------
# Clean Price
# -------------------------------
cars_df = cars_df[cars_df['Price'] != 'Ask For Price']

cars_df['Price'] = (
    cars_df['Price']
    .astype(str)
    .str.replace(',', '', regex=False)
)

cars_df['Price'] = pd.to_numeric(cars_df['Price'])

# -------------------------------
# Clean Kilometers Driven
# -------------------------------
cars_df = cars_df[cars_df['kms_driven'].notna()]

cars_df['kms_driven'] = (
    cars_df['kms_driven']
    .astype(str)
    .str.replace(' kms', '', regex=False)
    .str.replace(',', '', regex=False)
)

cars_df = cars_df[cars_df['kms_driven'].str.isnumeric()]
cars_df['kms_driven'] = cars_df['kms_driven'].astype(int)

# -------------------------------
# Clean Fuel Type
# -------------------------------
cars_df = cars_df[cars_df['fuel_type'].notna()]

# -------------------------------
# Remove Duplicates
# -------------------------------
cars_df.drop_duplicates(inplace=True)

# -------------------------------
# Reset Index
# -------------------------------
cars_df.reset_index(drop=True, inplace=True)

# -------------------------------
# Check Final Dataset
# -------------------------------
print(cars_df.info())
print(cars_df.head())
print(cars_df.shape)
````

**Question 1:**

Which car companies have the highest average selling price?

````python
import matplotlib.pyplot as plt
````

````python
import matplotlib.pyplot as plt

company_price = cars_df.groupby('company')['Price'].mean().sort_values(ascending=False).head(10)

company_price.plot(
    kind='bar',
    figsize=(10,5),
    color='#1F4E79'
)

plt.title('Top 10 Companies by Average Price')
plt.xlabel('Company')
plt.ylabel('Average Price (₹)')
plt.xticks(rotation=45)

plt.show()
````

**Ques2**. How has average car price changed over manufacturing years?

````python
year_price = cars_df.groupby('year')['Price'].mean()

year_price.plot(
    kind='line',
    figsize=(10,5),
    color='#1F4E79',
    marker='o'
)

plt.title('Average Price by Manufacturing Year')
plt.xlabel('Year')
plt.ylabel('Average Price (₹)')

plt.show()
````

**Ques 3**. Compare average price and number of listings by company

````python
company_count = cars_df['company'].value_counts().head(10)

company_avg_price = (
    cars_df.groupby('company')['Price']
    .mean()
    .loc[company_count.index]
)

fig, ax1 = plt.subplots(figsize=(10,5))

ax1.bar(
    company_count.index,
    company_count.values,
    color='#1F4E79'
)

ax1.set_ylabel('Number of Listings')

ax2 = ax1.twinx()

ax2.plot(
    company_avg_price.index,
    company_avg_price.values,
    color='orange',
    marker='o'
)

ax2.set_ylabel('Average Price')

plt.title('Listings vs Average Price')
plt.xticks(rotation=45)

plt.show()
````

**Ques 4**. What is the market share of each fuel type?

````python
import matplotlib.pyplot as plt

fuel_counts = cars_df['fuel_type'].value_counts()

# Navy Blue to Light Blue palette
colors = ['#1F4E79', '#2E75B6', '#5B9BD5', '#9DC3E6', '#D9EAF7']

plt.figure(figsize=(6,4))

fuel_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    shadow=True,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
    textprops={'fontsize': 11, 'fontweight': 'bold'}
)

plt.title(
    'Fuel Type Distribution',
    fontsize=18,
    fontweight='bold',
    pad=20
);

plt.ylabel('')  ;

plt.tight_layout();
plt.show();
````

**Ques 5** .What is the correlation between Year, Price and KMs Driven?

````python
import seaborn as sns

corr = cars_df[
    ['year','Price','kms_driven']
].corr()

plt.figure(figsize=(6,4))

sns.heatmap(
    corr,
    annot=True,
    cmap='Blues'
)

plt.title('Correlation Heatmap')

plt.show()
````
